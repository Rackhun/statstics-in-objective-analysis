- Monte Carlo methods, defined broadly, are also being used in ensemble weather and climate predictions, where multiple numerical forecasts are done using initial conditions/or forcings that vary within the range that they are known.
<br><br>
    
- One famous example is to calculate ${\pi}$. It can be calculated by inscribing a cricle within a square, then the ratio of areas between the circle and the square is ${\pi/4}$.
<br>
        <img src="./figure/monte_carlo.png" width="300">
<br><br>
    
- Now let drop points randomly and count how many points are in the circle (red). By dropping sufficiently large number of points, we can get an arbitrarily good approximation to ${\pi}$.  
<br><br>
    
    $${\pi  =  4\times\frac{\pi}{4}  =  4\times\frac{area \space of \space circle}{area \space of \space square}  \approx  4\times\frac{number \space of \space points \space in \space circle}{number \space of \space points \space of \space square}} $$
    

    
 ________________________________________________________________________________________________

```python
import random
import matplotlib.pyplot as plt

random.seed(4)

out_circle_x, out_circle_y = [], []
in_circle_x,  in_circle_y  = [], []

for _ in range(1000):
    x = random.uniform(-1,1)
    y = random.uniform(-1,1)
    
    if x**2 + y**2 > 1:
        out_circle_x.append(x)
        out_circle_y.append(y)
    else:
        in_circle_x.append(x)
        in_circle_y.append(y)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(out_circle_x, out_circle_y, color='blue')
plt.scatter(in_circle_x,  in_circle_y, color='red')

approximated_pi = 4 * len(in_circle_x) / (len(out_circle_x) + len(in_circle_x))
plt.title(f"approximated pi is {approximated_pi} from 1000 samples", fontsize=20)

plt.show()
```
